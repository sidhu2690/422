Here we have to compare three different computational methods that should all produce the same sequence (1/3^n), but they start with slightly wrong initial values

```python
import matplotlib.pyplot as plt
```

```python
n_value = 50  #iterations

true_sequence = [(1/3)**n for n in range(n_value)]

r = [0.99996]
for n in range(1, n_value):
  r.append(1/3 * r[n-1])


error_r = [abs( r[n] - true_sequence[n] ) for n in range(n_value)]



n_vals = [n for n in range(n_value)]
```

```python
plt.figure(figsize = (5, 4))
plt.semilogy(n_vals, error_r)
plt.xlabel("n_values")
plt.ylabel("Absolute Error (log scale)")
plt.show()
```

```python
p = [1, 0.33332]

for n in range(2, n_value):
  p.append((4/3) * p[n-1] - (1/3) * p[n-2])

error_p = [abs( p[n] - true_sequence[n] ) for n in range(n_value)]

plt.figure(figsize = (5, 4))
plt.semilogy(n_vals, error_p)
plt.xlabel("n_values")
plt.ylabel("Absolute Error (log scale)")
plt.show()
```

```python
q = [1, 0.33332]

for n in range(2, n_value):
  q.append((10/3) * q[n-1] - q[n-2])

error_q = [abs( q[n] - true_sequence[n] ) for n in range(n_value)]

plt.figure(figsize = (5, 4))
plt.semilogy(n_vals, error_q)
plt.xlabel("n_values")
plt.ylabel("Absolute Error (log scale)")
plt.show()
```
