```python
def row_swap(matrix, i, j):
  matrix = np.array(matrix)
  matrix[[i, j]] = matrix[[j, i]]
  return matrix
```

```python
A = [[10, 20, 30],
     [40, 50, 60],
     [70, 80, 90]]

row_swap(A, 0, 2)
```

```python
def row_scale(matrix, i, alpha):
  matrix = np.array(matrix)
  matrix[i] *= alpha
  return matrix
```

```python
A = [[10, 20, 30],
     [40, 50, 60],
     [70, 80, 90]]

row_scale(A, 0, 5)
```

```python
def row_add(matrix, i, j, alpha):
  matrix = np.array(matrix)
  matrix[i] = matrix[i] * alpha + matrix[j]
  return matrix
```

```python
A = [[10, 20, 30],
     [40, 50, 60],
     [70, 80, 90]]

row_add(A, 0, 1, 5)
```
