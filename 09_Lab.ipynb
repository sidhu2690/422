```python
def composite_trapezoidal(a, b, n):
  h = (b - a) / n
  x = np.linspace(a, b, n + 1)
  y = f(x)
  I = (h / 2) * (y[0] + 2 * np.sum(y[1:-1]) + y[-1])
  return I
```

```python
def f(x):
  return x**2 + 1

composite_trapezoidal(0, 2, 4).item()
```

```python
def composite_simpson(a, b, m):
  if m % 2 != 0:
    raise ValueError("n must be even for Simpson's Rule")
  h = (b - a) / m
  x = np.linspace(a, b, m + 1)
  y = f(x)
  I = (h / 3) * (y[0] + 4 * np.sum(y[1:-1:2]) + 2 * np.sum(y[2:-2:2]) + y[-1])
  return I
```

```python
composite_simpson(0, 2, 4).item()
```

```python
def gauss_legendre_quadrature(a, b, N):
  wk, xk = zip(*data[N])
  wk = np.array(wk)
  xk = np.array(xk)

  t = 0.5 * (a + b) + 0.5 * (b - a) * xk
  ft = f(t)

  I = 0.5 * (b - a) * np.sum(wk * ft)
  return I


gauss_legendre_quadrature(1, 6, N=3).item() - I
```
